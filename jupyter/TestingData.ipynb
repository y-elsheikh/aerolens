```python
import datetime
today = datetime.date.today()
yesterday = today - datetime.timedelta(days=1)
```

```python
today.isoformat()
```

```python
san_francisco = (-122.536285,37.700349,-122.357071,37.808325)

no2data = ea.search_data(
    short_name="TEMPO_NO2_L3",
    temporal=(yesterday.isoformat(),today.isoformat()),
    bounding_box=san_francisco
)
assert tempo_data, "No granules found"

ozonedata = ea.search_data(
    short_name="TEMPO_O3TOT_L3_V03",
    temporal=(yesterday.isoformat(),today.isoformat()),
    bounding_box=san_francisco
)
assert tempo_data, "No granules found"

formaldehydedata = ea.search_data(
    short_name="TEMPO_HCHO_L3",
    temporal=(yesterday.isoformat(),today.isoformat()),
    bounding_box=san_francisco
)
assert tempo_data, "No granules found"

clouddata = ea.search_data(
    short_name="TEMPO_CLDO4_L3",
    temporal=(yesterday.isoformat(),today.isoformat()),
    bounding_box=san_francisco
)
assert tempo_data, "No granules found"
```

```python
try:
  ea.download(no2data, local_path="data/TEMPO_NO2_L3_Data/")
except ValueError:
  print("No NO2 Data")
  pass
try:
  ea.download(ozonedata, local_path="data/TEMPO_O3TOT_L3_V03_Data/")
except ValueError:
  print("No Ozone Data")
  pass
try:
  ea.download(formaldehydedata, local_path="data/TEMPO_HCHO_L3_Data/")
except ValueError:
  print("No Formaldehyde Data")
  pass
try:
  ea.download(clouddata, local_path="data/TEMPO_CLDO4_L3_Data/")
  pass
except ValueError:
  print("No Cloud Data")
```
